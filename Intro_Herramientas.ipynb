# Como acceder a documentos o archivos desde Colab


## Cargando Archivos

Cargar los archivos desde la maquina local hasta la nube


```python
from google.colab import files

uploaded_2 = files.upload()
```

```python
print(uploaded_2.keys())
```

```python
print(uploaded_2)
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_byte_array = uploaded_2['scikit.png']
img_arr = np.frombuffer(img_byte_array, np.uint8)
img = cv2.imdecode(img_arr, cv2.IMREAD_COLOR)
```

```python
plt.imshow(img)
```

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

## Como descargar scripts o archivos

Se pueden enviar documentos de la maquina en la nube a la maquina local

```python
with open('example.txt', 'w') as f:
  f.write('Hola Mundo')
  f.write('test')

f = open("example_2.txt", 'w')
f.writelines(["hola", "test"])
f.close()
```

```python
from google.colab import files

files.download('example_2.txt')
```
